```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# 크롬 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL에 접속
url = 'https://maplescouter.com/total-ranking?job=%EC%A0%84%EC%B2%B4&world=%EC%A0%84%EC%B2%B4&guild=&page=1&pageSize=30'
driver.get(url)

# 페이지가 완전히 로드될 때까지 기다림
time.sleep(5)

# CSV 파일을 열어 데이터를 저장할 준비
csv_file = open('C:/MapleStory_rank/헥사랭킹.csv', mode='w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)

# CSV 헤더 작성
csv_writer.writerow(['순위', '닉네임', '레벨', '직업', '길드', '아이템 환산', '헥사 환산'])

# 데이터 추출을 위한 반복
page_num = 1
while True:  # 계속해서 페이지를 넘기며 반복
    # 페이지 소스 가져오기
    page_source = driver.page_source

    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(page_source, 'html.parser')

    # 순위, 닉네임, 레벨, 직업, 길드, 아이템 환산, 헥사 환산 추출
    rankings = soup.find_all('tr', class_='text-center font-normal hover:bg-[#E8F1E9]')

    # 랭킹 데이터 추출
    if rankings:
        for ranking in rankings:
            try:
                # 순위
                rank = ranking.find('div').get_text(strip=True)

                # 닉네임
                nickname = ranking.find('a').get_text(strip=True)

                # 레벨
                level = ranking.find_all('td')[2].get_text(strip=True)

                # 직업
                job = ranking.find_all('td')[3].get_text(strip=True)

                # 길드
                guild = ranking.find_all('td')[4].get_text(strip=True)

                # 아이템 환산
                item_conversion = ranking.find_all('td')[5].get_text(strip=True)

                # 헥사 환산
                hex_conversion = ranking.find_all('td')[6].get_text(strip=True)

                # 헥사 환산이 100,000 이상인 경우만 CSV에 저장
                if int(hex_conversion.replace(',', '')) >= 100000:
                    csv_writer.writerow([rank, nickname, level, job, guild, item_conversion, hex_conversion])

                # 헥사 환산이 100,000 미만인 경우 반복 종료
                else:
                    print("헥사 환산 100,000 미만 유저 발견. 데이터 추출 종료.")
                    csv_file.close()
                    driver.quit()
                    exit()  # 프로그램 종료

            except AttributeError as e:
                print("데이터 추출 중 오류 발생:", e)
    else:
        print("랭킹 데이터가 없습니다.")
        break

    # 'Next Page' 버튼 클릭하여 다음 페이지로 이동
    try:
        next_button = driver.find_element(By.XPATH, '//li[@title="Next Page"]/button')
        next_button.click()
        page_num += 1
        print(f"Page {page_num}로 이동 중...")
        time.sleep(5)  # 페이지가 완전히 로드될 때까지 기다림
    except Exception as e:
        print("다음 페이지로 이동할 수 없습니다. 또는 마지막 페이지에 도달했습니다.")
        break

# 브라우저 종료
driver.quit()
```
